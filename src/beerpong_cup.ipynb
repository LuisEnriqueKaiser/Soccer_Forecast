```python
import pandas as pd
import random

def team_to_number_random(seed, teams, team_names):
    random.seed(seed)
    team_to_number = {}
    for team in teams:
        team_to_number[team] = random.choice(team_names)
        team_names.remove(team_to_number[team])
    return team_to_number

```
